```python
import time
import numpy as np


matrix_A = [[1, 2], [3, 4]]
matrix_B = [[5, 6], [7, 8]]
result_matrix = [[0, 0], [0, 0]]

start_time = time.time()
for i in range(len(matrix_A)):
    for j in range(len(matrix_B[0])):
        for k in range(len(matrix_B)):
            result_matrix[i][j] += matrix_A[i][k] * matrix_B[k][j]
end_time = time.time()

print("Result from loops:", result_matrix)
print(f"Time taken: {end_time - start_time:.6f} seconds (instant)")
```

# Loops vs numpy

```python
size = 500

matrix_a_list = [[i + j for j in range(size)] for i in range(size)]
matrix_b_list = [[i - j for j in range(size)] for i in range(size)]

# ----------- Python Loop Version -----------
result_list = [[0 for _ in range(size)] for _ in range(size)]

start_time = time.time()
for i in range(size):
    for j in range(size):
        total = 0
        for k in range(size):
            total += matrix_a_list[i][k] * matrix_b_list[k][j]
        result_list[i][j] = total
end_time = time.time()
python_time = end_time - start_time
print(f"\nPython loop took: {python_time:.4f} seconds")

# ----------- NumPy Version -----------
matrix_a_np = np.array(matrix_a_list)
matrix_b_np = np.array(matrix_b_list)

start_time = time.time()
result_np = np.dot(matrix_a_np, matrix_b_np)
end_time = time.time()
numpy_time = end_time - start_time
print(f"NumPy took: {numpy_time:.4f} seconds")

# ----------- Comparison -----------
speedup = python_time / numpy_time if numpy_time != 0 else float('inf')
print(f"\nCONCLUSION: NumPy was approximately {int(speedup)}Ã— faster!")
```
